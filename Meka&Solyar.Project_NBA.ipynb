```python
misc2 = misc1.drop(['W','Attend.','MOV','SRS','L','TS%','FTr','DRtg','PW','PL','ORtg','OppeFG%','eFG%'], axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
trad2.columns
```

```python
X = trad2[['FG%', '3P', '3P%', '2P%', 'FT', 'FT%', 'ORB', 'DRB', 'STL', 'BLK', 'TOV', 'PF', 'PTS']]
Y = trad2["W/L%"]
X_train1, X_val1, Y_train1, Y_val1 = train_test_split(X,Y,test_size=0.25)
model = LinearRegression()
model.fit(X_train1, Y_train1)
Y_train_predicted1 = model.predict(X_train1)
Y_val_predicted1 = model.predict(X_val1)
train_error1 = mean_squared_error(Y_train_predicted1, Y_train1)
val_error1 = mean_squared_error(Y_val_predicted1, Y_val1)
print('ÐžÑˆÐ¸Ð±ÐºÐ° Ð½Ð° Ð¾Ð±ÑƒÑ‡ÐµÐ½Ð¸Ð¸:  \t', train_error1)
print('ÐžÑˆÐ¸Ð±ÐºÐ° Ð½Ð° Ð²Ð°Ð»Ð¸Ð´Ð°Ñ†Ð¸Ð¸:\t', val_error1)
print('Ð Ð°Ð·Ð½Ð¸Ñ†Ð°:\t\t', val_error1 - train_error1)
```
